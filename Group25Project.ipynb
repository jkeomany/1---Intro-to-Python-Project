```python
import pandas as pd
import numpy as np

# Canadian Data 
# Import and clean the data set
df_can = pd.read_csv("can_data.csv")

## Make a list of columns to drop
drop_list = ['DGUID','Statistic','UOM','UOM_ID','SCALAR_FACTOR','SCALAR_ID','VECTOR','COORDINATE','STATUS','SYMBOL','TERMINATED','DECIMALS']
## The columns in drop list were removed from the statistical analysis because the columns did not provide value for the regression model and was not informative.

## Make the Canadian household expenditures dataframe
df_can = df_can.drop(drop_list, axis=1)

## Rename long and unclear column names
df_can.rename(columns={"Household expenditures, summary-level categories": "Household expenditures"}, inplace=True)

df_can.head()
```

```python
# US Data https://www.bls.gov/cex/tables.htm
# Import and clean the data set
df_us = pd.read_excel("cu-all-multi-year-2013-2020.xlsx",header=0)

## Replace the header with the year data
new_header = df_us.iloc[1] # Select where the year data row
df_us = df_us.iloc[2:] # Replace the dataframe with 2 headers removed
df_us.columns = new_header # Set the new header
df_us = df_us.iloc[43:] # Remove consumer characteristics and percent distribution data

## Drop NA values
df_us = df_us.dropna()
df_us = df_us.reset_index(drop=True)

## Remove the Item column
df_us = df_us.drop('Item',axis=1)

df_us.head()
```

```python
# US Data 2
# Import and clean the data set
df_us2 = pd.read_excel("cu-all-multi-year-2006-2012.xlsx",header=0)

## Replace the header with the year data
new_header = df_us2.iloc[1] # Select where the year data row
df_us2 = df_us2.iloc[2:] # Replace the dataframe with 2 headers removed

## Fix the header
new_header[3:6] = 2800, 2009, 2010 # Removes the decimals from columns
df_us2.columns = new_header # Set the new header
df_us2 = df_us2.iloc[43:] # Remove consumer characteristics and percent distribution data

## Drop NA
df_us2 = df_us2.dropna()
df_us2 = df_us2.reset_index(drop=True)

## Merge df_us and df_us2
merged_df = pd.concat([df_us2,df_us],axis=1)

merged_df.head()
```
